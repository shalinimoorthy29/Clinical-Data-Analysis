### Cleaning and Inspecting Survival Data

**Questions Addressed:**

- How can we clean and prepare survival data (Overall Survival Days and Overall Survival Status) for analysis?
- Are there any missing or invalid values in these survival-related columns, and how should they be handled?

**Why Clean Survival Data?**

Survival analysis relies on clean and complete data for accurate results.  
Key variables include:  
- **Overall Survival Days**: Total days a patient survived from diagnosis.  
- **Overall Survival Status**: Binary outcome indicating whether the patient is **LIVING** or **DECEASED**.  

Ensuring these columns are free of missing or invalid values is a crucial first step in survival modelling.

We remove rows with missing `Overall Survival Days` because survival time is critical for analysis.  
Using `subset` ensures that only this column is checked, leaving other data untouched.


```python
# Filter Data by Overall Survival Days
clinical_data_copy_os_days = clinical_data_copy.dropna(subset=['Overall Survival Days'])

# Check the shape of the cleaned DataFrame for overall survival days
print(f"Shape after cleaning 'Overall Survival Days': {clinical_data_copy_os_days.shape}")

# Display the first few rows if you want to inspect the cleaned data
print("First few rows of 'Overall Survival Days' and 'Overall Survival Status':")
print(clinical_data_copy_os_days[['Overall Survival Days', 'Overall Survival Status']].head(11))

# Check for missing values in these columns
print("\nMissing values in 'Overall Survival Days' and 'Overall Survival Status' after reloading:")
print(clinical_data_copy_os_days[['Overall Survival Days', 'Overall Survival Status']].isnull().sum())
```

```python
# Extract numeric part from 'Overall Survival Status' by splitting at ':'
clinical_data_copy_os_days.loc[:, 'Overall Survival Status'] = clinical_data_copy_os_days['Overall Survival Status'].apply(
    lambda x: int(str(x).split(':')[0]) if pd.notnull(x) else x
)

# Convert 'Overall Survival Days' to numeric with coercion
clinical_data_copy_os_days.loc[:, 'Overall Survival Days'] = pd.to_numeric(clinical_data_copy_os_days['Overall Survival Days'], errors='coerce')

# Convert 'Overall Survival Status' to numeric with coercion
clinical_data_copy_os_days.loc[:, 'Overall Survival Status'] = pd.to_numeric(clinical_data_copy_os_days['Overall Survival Status'], errors='coerce')

# Drop rows with missing values in the survival columns
survival_data = clinical_data_copy_os_days.dropna(subset=['Overall Survival Days', 'Overall Survival Status'])

# Verify the cleaned data
print("First few rows after cleaning:")
print(survival_data[['Overall Survival Days', 'Overall Survival Status']].head())

print("\nRemaining missing values in survival columns after cleaning:")
print(survival_data[['Overall Survival Days', 'Overall Survival Status']].isnull().sum())
```

### Kaplan-Meier Survival Analysis – Overall Survival

**Question Addressed:**

What is the overall survival probability for patients in the dataset, and how does it change over time?

```python
import matplotlib.pyplot as plt
from lifelines import KaplanMeierFitter

# Define time to event (T) and event occurrence (E)
T = survival_data['Overall Survival Days'] / 30  # Time in months
E = survival_data['Overall Survival Status']  # Event indicator (1 = event occurred, 0 = censored)


# Initialise the Kaplan-Meier fitter
kmf = KaplanMeierFitter()

# Fit the data
kmf.fit(durations=T, event_observed=E, label="Overall Survival")

# Plot the survival function
plt.figure(figsize=(10, 6))
kmf.plot_survival_function()
plt.title("Kaplan-Meier Survival Curve")
plt.xlabel("Months")  # Update the x-axis label to "Months"
plt.ylabel("Survival Probability")
plt.xlim(0, 140)  # Adjust x-axis limits for better visibility
plt.grid()
plt.show()
```

**Conclusion:**

The Kaplan-Meier survival curve illustrates the overall survival probability for all patients as a single group. Key observations include:

- Survival Decline: - At **12 months**, the survival probability is approximately **75%**.  The survival probability decreases steadily over time from 20 months onwards, reflecting the progression of the disease and treatment outcomes across the population.

- Confidence Interval: The shaded region represents the **95% confidence interval** around the estimated survival probabilities. A wider interval reflects greater uncertainty due to fewer patients at risk at later time points.

This analysis offers a comprehensive overview of survival trends in the dataset but does not provide subgroup-specific insights (e.g., based on mutations or demographics), which may be explored in subsequent analyses.

### Relationship between mutation status and survival

### Kaplan-Meier Survival Curve by FLT3-ITD Status

**Questions Addressed:**

- Does the presence of the FLT3-ITD mutation impact overall survival probability in the dataset?
- Is the difference in survival between FLT3-ITD positive and negative patients statistically significant?

```python
# Standardise values in 'FLT3-ITD Postive' to uppercase
survival_data['FLT3-ITD Postive'] = survival_data['FLT3-ITD Postive'].str.upper()

# Exclude rows where 'FLT3-ITD Postive' is 'UNKNOWN'
filtered_data = survival_data[survival_data['FLT3-ITD Postive'] != 'UNKNOWN']

# Define time to event and event indicator
T = filtered_data['Overall Survival Days'] / 30  # Convert days to months
E = filtered_data['Overall Survival Status']     # Event indicator

# Plot the survival function by standardized 'FLT3-ITD Postive' status
plt.figure(figsize=(10, 6))
for status in filtered_data['FLT3-ITD Postive'].dropna().unique():
    kmf.fit(T[filtered_data['FLT3-ITD Postive'] == status], E[filtered_data['FLT3-ITD Postive'] == status], label=f"FLT3-ITD Postive: {status}")
    kmf.plot_survival_function()

plt.title("Kaplan-Meier Survival Curve by FLT3-ITD Mutation Status")
plt.xlabel("Months")
plt.ylabel("Survival Probability")
plt.xlim(0, 140)
plt.grid()
plt.legend(title="FLT3-ITD Status")
plt.show()
```

```python
from lifelines.statistics import logrank_test

# Define time and event indicator for each group
T_yes = filtered_data[filtered_data['FLT3-ITD Postive'] == 'YES']['Overall Survival Days'] / 30  # Time in months
E_yes = filtered_data[filtered_data['FLT3-ITD Postive'] == 'YES']['Overall Survival Status']

T_no = filtered_data[filtered_data['FLT3-ITD Postive'] == 'NO']['Overall Survival Days'] / 30
E_no = filtered_data[filtered_data['FLT3-ITD Postive'] == 'NO']['Overall Survival Status']

# Perform the log-rank test
results = logrank_test(T_yes, T_no, event_observed_A=E_yes, event_observed_B=E_no)

# Check if the result is significant and print a message
if results.p_value < 0.05:
    print("The difference in survival between FLT3-ITD Positive (YES) and (NO) groups is statistically significant (p-value: {:.4f}).".format(results.p_value))
else:
    print("The difference in survival between FLT3-ITD Positive (YES) and (NO) groups is not statistically significant (p-value: {:.4f}).".format(results.p_value))

# Print the summary of the test
results.print_summary()
```

**Conclusions:**

- Kaplan-Meier Survival Curve:

Patients with **FLT3-ITD Positive** status show a **steeper decline** in survival probability within the first **40 months** compared to FLT3-ITD Negative patients. By **60 months**, survival stabilises for both groups, but FLT3-ITD Positive patients remain at a consistently lower survival probability. Patients with FLT3-ITD positive status exhibit lower survival probabilities over time compared to FLT3-ITD negative patients. The separation of survival curves suggests that FLT3-ITD mutation status significantly impacts survival outcomes, with poorer prognosis for positive patients. FLT3-ITD mutations are a known negative prognostic marker in AML. Patients carrying this mutation often require **more aggressive treatment** and closer monitoring. These results reinforce its role in risk stratification and treatment decision-making.

- Statistical Test (Log-Rank):

A log-rank test was performed to assess the statistical significance of the survival difference between the two groups.
The p-value (0.0092) indicates that the difference is statistically significant, confirming that FLT3-ITD status has a meaningful impact on survival outcomes.

- Interpretation:
 
The analysis demonstrates that the presence of the FLT3-ITD mutation is associated with worse overall survival in paediatric AML patients. This finding highlights the clinical importance of FLT3-ITD as a prognostic biomarker and reinforces the need to consider mutation status in treatment planning and risk stratification.

Groups with small sample sizes (**n ≤ 50**) were excluded to reduce bias. Small group sizes can produce unstable survival estimates and inflated confidence intervals, which may distort pairwise comparisons.

```python
import itertools
from lifelines.statistics import logrank_test

# Filter out race categories with sample sizes <= 50 for statistical reliability
filtered_data = filtered_data[filtered_data['Race Category'].map(race_counts) > 50]

# Filter T and E based on the filtered data
T = filtered_data['Overall Survival Days'] / 30  # Convert days to months
E = filtered_data['Overall Survival Status']

# Re-run the pairwise log-rank test
race_categories = filtered_data['Race Category'].unique()
p_values = {}

for (race1, race2) in itertools.combinations(race_categories, 2):
    group1 = filtered_data['Race Category'] == race1
    group2 = filtered_data['Race Category'] == race2
    results = logrank_test(
        T[group1], T[group2],  # Filtered T
        event_observed_A=E[group1], event_observed_B=E[group2]  # Filtered E
    )
    p_values[f"{race1} vs {race2}"] = results.p_value

# Convert p-values to DataFrame
p_values_df = pd.DataFrame.from_dict(p_values, orient='index', columns=['p-value'])
p_values_df = p_values_df.sort_values(by='p-value')
p_values_df
```

- Filtering by Group Size:

To ensure the statistical robustness of the analysis, groups with small sample sizes (n ≤ 50) were excluded. Accounting for variation in counts is essential because small sample sizes can lead to unstable survival estimates and inflated confidence intervals, potentially biasing the results.

- Statistical Test (Log-Rank):

### Kaplan-Meier Survival Analysis by Minimal Residual Disease (MRD) Groups

**Question Addressed:**

- Does the level of Minimal Residual Disease (MRD Percentage at the end of Course 1) influence survival outcomes?
- Are there statistically significant differences in survival between High MRD and Low MRD groups?


```python
# Check for missing values in the column
missing_mrd = survival_data['MRD Percentage at end of Course 1'].isnull().sum()
print(f"Number of missing values in 'MRD Percentage at end of Course 1': {missing_mrd}")
```

```python
survival_data = survival_data.dropna(subset=['MRD Percentage at end of Course 1'])
```

```python
import numpy as np

# Define high and low MRD groups
median_mrd = survival_data['MRD Percentage at end of Course 1'].median()
survival_data['MRD Group'] = np.where(
    survival_data['MRD Percentage at end of Course 1'] > median_mrd, 
    'High MRD', 
    'Low MRD'
)
```

```python
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt

# Define survival times and event indicators
T = survival_data['Overall Survival Days'] / 30  # Convert days to months
E = survival_data['Overall Survival Status']  # 1 = event occurred, 0 = censored

# Initialise the Kaplan-Meier Fitter
kmf = KaplanMeierFitter()

# Plot survival curves for each MRD group
plt.figure(figsize=(10, 6))
for group in survival_data['MRD Group'].unique():
    mask = survival_data['MRD Group'] == group
    kmf.fit(T[mask], event_observed=E[mask], label=group)
    kmf.plot_survival_function()

# Add labels and title
plt.title('Kaplan-Meier Survival Curve by MRD Group')
plt.xlabel('Months')
plt.ylabel('Survival Probability')
plt.grid()
plt.show()
```

```python
from lifelines.statistics import logrank_test

# Separate the data into High MRD and Low MRD groups
high_mrd = survival_data[survival_data['MRD Group'] == 'High MRD']
low_mrd = survival_data[survival_data['MRD Group'] == 'Low MRD']

# Perform the log-rank test
results = logrank_test(
    high_mrd['Overall Survival Days'] / 30,  # Time in months for High MRD group
    low_mrd['Overall Survival Days'] / 30,   # Time in months for Low MRD group
    event_observed_A=high_mrd['Overall Survival Status'],  # Event indicator for High MRD
    event_observed_B=low_mrd['Overall Survival Status']    # Event indicator for Low MRD
)

# Print the log-rank test results
print('Log-Rank Test Results:')
print(f' - Test Statistic: {results.test_statistic:.4f}')
print(f' - p-value: {results.p_value:.2e}')  # Display p-value in scientific notation

# Interpretation based on p-value
if results.p_value < 0.05:
    print("The survival difference between High MRD and Low MRD groups is statistically significant.")
else:
    print("The survival difference between High MRD and Low MRD groups is not statistically significant.")
```

**Conclusions:**

- Kaplan-Meier Analysis: 

**Low MRD** patients exhibit significantly **higher survival probabilities** compared to **High MRD** patients. By **60 months**, survival for the Low MRD group stabilises at ~70%, whereas survival for the High MRD group drops below 50%. The survival difference is particularly pronounced within the first **24 months**, indicating the early impact of MRD levels.

- Statistical Test (Log-Rank Test): 

The log-rank test confirmed that the difference in survival between High MRD and Low MRD groups is **highly significant** significant (p-value = 3.06e-14).

- Clinical Implications: 

Minimal Residual Disease (MRD) levels strongly predict survival outcomes in paediatric AML patients. **High MRD** is associated with **poorer survival**, likely due to residual disease persistence post-treatment. **Low MRD** reflects a better treatment response and improved prognosis. These findings reinforce the importance of MRD as a **prognostic biomarker** for identifying patients at high risk of relapse and guiding therapeutic strategies.

### Log-Rank Test for Survival Differences Across Age Groups

```python
from lifelines.statistics import logrank_test
import itertools

# Ensure the DataFrame is a proper copy to avoid SettingWithCopyWarning
age_data = clinical_data_copy.dropna(subset=['Diagnosis Age']).copy()

# Create age groups based on 'Diagnosis Age'
age_data['Age Group'] = pd.cut(
    age_data['Diagnosis Age'],
    bins=[0, 5, 10, 15, 20, 25, 30],
    labels=['<5', '5-10', '10-15', '15-20', '20-25', '>25']
)

# Display the shape of the cleaned DataFrame and inspect the 'Age Group' column
print(f"Shape of the dataset after adding 'Age Group': {age_data.shape}")
print(age_data[['Diagnosis Age', 'Age Group']].head())

# Perform log-rank tests for each pair of age groups
age_groups = age_data['Age Group'].dropna().unique()
p_values = {}

for (age1, age2) in itertools.combinations(age_groups, 2):
    group1 = age_data['Age Group'] == age1
    group2 = age_data['Age Group'] == age2
    results = logrank_test(
        age_data.loc[group1, 'Overall Survival Days'] / 30,  # Convert days to months
        age_data.loc[group2, 'Overall Survival Days'] / 30,
        event_observed_A=age_data.loc[group1, 'Overall Survival Status'],
        event_observed_B=age_data.loc[group2, 'Overall Survival Status']
    )
    p_values[f"{age1} vs {age2}"] = results.p_value

# Display p-values for statistical significance between age groups
p_values_df = pd.DataFrame.from_dict(p_values, orient='index', columns=["p-value"])
p_values_df
```

- Survival Trends Across Age Groups:

Patients were categorised into age groups: <5, 5–10, 10–15, 15–20, 20–25, and >25 years. Pairwise log-rank tests were conducted to assess survival differences between these groups.

Significant survival differences were observed:

    - Between 5–10 years and 15-20 years (p = 0.000552).

- Key Observations:

- Interpretation:

### MRD Percentage by Risk Group

**Question Addressed:**

- Are there significant differences in MRD percentages across the different Risk Groups?

```python
# Boxplot of MRD Percentage by Risk Group
plt.figure(figsize=(10, 6))
sns.boxplot(data=survival_data, x='Risk Group', y='MRD Percentage at end of Course 1', palette='Set2')
plt.title("MRD Percentage by Risk Group")
plt.xlabel("Risk Group")
plt.ylabel("MRD Percentage")
plt.grid()
plt.show()
```

```python
from scipy.stats import kruskal

# Extract MRD Percentage values for each Risk Group
low_risk_mrd = survival_data[survival_data['Risk Group'] == 'Low']['MRD Percentage at end of Course 1'].dropna()
standard_risk_mrd = survival_data[survival_data['Risk Group'] == 'Standard']['MRD Percentage at end of Course 1'].dropna()
high_risk_mrd = survival_data[survival_data['Risk Group'] == 'High']['MRD Percentage at end of Course 1'].dropna()

# Perform Kruskal-Wallis H-test
kruskal_stat, kruskal_p_value = kruskal(low_risk_mrd, standard_risk_mrd, high_risk_mrd)

print(f"Kruskal-Wallis H-test statistic: {kruskal_stat:.4f}")
print(f"P-value: {kruskal_p_value:.4e}")

# Interpretation
if kruskal_p_value < 0.05:
    print("The differences in MRD percentages between the Risk Groups are statistically significant.")
else:
    print("The differences in MRD percentages between the Risk Groups are not statistically significant.")
```

**Conclusions:**

- Boxplot Insights:

The boxplot highlights noticeable differences in MRD percentages among the Risk Groups. The High Risk Group demonstrates markedly higher MRD percentages compared to the Low and Standard Risk Groups.

- Statistical Test Results (Kruskal-Wallis H-test):

The Kruskal-Wallis test confirmed that these differences are statistically significant (H-statistic = 60.27, p-value = 8.17e-14).
This indicates that MRD percentages vary significantly across Risk Groups.

- Interpretation:
The observed variations in MRD percentages align with risk stratification, as the High Risk Group exhibits elevated MRD values. This highlights the potential importance of MRD in evaluating treatment response and risk classification for clinical decision-making.
